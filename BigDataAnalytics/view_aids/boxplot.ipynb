```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Function to plot a sine wave with adjustable amplitude and frequency
def plot_sine_wave(amplitude=1.0, frequency=1.0):
    x = np.linspace(0, 2 * np.pi, 400)
    y = amplitude * np.sin(frequency * x)
    
    plt.figure(figsize=(8, 4))
    plt.plot(x, y)
    plt.ylim(-10, 10)
    plt.title(f"Sine Wave: Amplitude={amplitude:.2f}, Frequency={frequency:.2f}")
    plt.xlabel("x")
    plt.ylabel("sin(x)")
    plt.grid(True)
    plt.show()

# Create interactive sliders for amplitude and frequency
interact(plot_sine_wave, 
         amplitude=FloatSlider(value=1.0, min=0.1, max=10.0, step=0.1), 
         frequency=FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1))
```
